**Работа с API**

Импортируем бибилиотеки

```python
import requests
import pandas as pd
```

Получить информацию обо всех книгах и сохранить в pd.DataFrame

```python

books_url = "https://www.anapioficeandfire.com/api/books"
books = requests.get(books_url).json()
books_df = pd.DataFrame(books)

display(books_df.head())
```

Получить информацию обо всех домах Вестероса (Houses) и сохранить в pd.DataFrame

```python
all_houses = []

url = f"https://anapioficeandfire.com/api/houses?region=The Westerlands"
response = requests.get(url)
data = response.json()
all_houses.extend(data)


df = pd.DataFrame(all_houses)
print(df[["name", "region", "words"]].head())
```

Получить информацию обо всех домах Вестероса, у которых есть девиз, и сохранить в pd.DataFrame

```python
all_houses_with_words = []

url = f"https://anapioficeandfire.com/api/houses?region=The Westerlands&hasWords=true"
response = requests.get(url)
data = response.json()

all_houses_with_words.extend(data)


df_westerlands_words = pd.DataFrame(all_houses_with_words)
print(df_westerlands_words[["name", "region", "words"]].head(15))
```
